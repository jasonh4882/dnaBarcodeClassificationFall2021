```python
import sklearn.ensemble
```

```python
clf = sklearn.ensemble.RandomForestClassifier(n_jobs=-1, max_features="auto", n_estimators=1000, criterion="gini")

clf.fit(bulk_train_features, nuc_bulk_train_label)
```

```python
reserve_predict =clf.predict(reserve_train_features)
```

```python
correct = 0
lngth = len(nuc_reserve_train_label)
for row in range(0, lngth):
    
    compared = (nuc_reserve_train_label[row] == reserve_predict[row])
    if compared:
        correct += 1
    else:
        print("{}:{} ?= {}: {}".format(row, nuc_reserve_train_label[row], reserve_predict[row], compared))
print("Accuracy: {}/{} [{}%]".format(correct, lngth, (correct/lngth)*100)) 
```

```python
both_features = [bulk_train_features,reserve_train_features]
both_labels = [nuc_bulk_train_y,nuc_reserve_train_y]
all_train_features = pd.concat(both_features)
both_labels_conc = pd.concat(both_labels)
all_train_labels = both_labels_conc['labels']
```

```python
finalRF = sklearn.ensemble.RandomForestClassifier(n_jobs=10, random_state=0)

finalRF.fit(all_train_features, all_train_labels)
```

```python
test_predict =finalRF.predict(test_features)
```
